```python
def my_plot(x,y,nr):
    plt.subplot(3,2,nr)
    sns.countplot(x=df[x])
    plt.title(x+' counting')

    plt.subplot(3, 2, nr+1)
    sns.boxplot(x=df[x], y=df[y])
    plt.title(x + ' vs Price')


plt.figure(figsize=(20,15))
my_plot('transmission','price',1)
my_plot('fuelType','price',3)
```

1.The most sold cars are those with manual transmission. (from 1st graph)
2.Cars with the highest price are those with Automatic and Semi-auto transmissions. (from 2nd graph)
3.The most used fuel is petrol followed by diesel. (from 3rd graph)
4.Cars with Hybrid fueltype have the highest price. (from 4th graph)

```python
x=df.drop("price",axis=1) #splitting data in x and y
y=df["price"]
```

```python
print("Shape of x is : ",x.shape)
print("shape of y is : ",y.shape)
```

```python
from sklearn.model_selection import train_test_split # splitting in training and testing data
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print("Shape of xtrain is : ",xtrain.shape)
print("shape of ytrain is : ",ytrain.shape)
print("Shape of xtest is : ",xtest.shape)
print("shape of ytest is : ",ytest.shape)
```

```python
from sklearn.preprocessing import StandardScaler #normalizing data using scalar
ss=StandardScaler()  
xtrain=ss.fit_transform(xtrain)
xtest=ss.transform(xtest)
```

```python
df.head()
```

```python
#applying different models on data
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor()
```

```python
regression_models = [lr,rfr,gbr,dtr,xgb]
score_train = list()
score_test = list()

for model in regression_models : 
    model.fit(xtrain,ytrain)
    y_pred = model.predict(xtest)
    
    score_train.append(model.score(xtrain,ytrain))
    score_test.append(model.score(xtest,ytest))
```

```python
model_names = ['Linear Regression','Random Forest Regressor','Gradient Boosting Regressor','Decision Tree Regressor','XGBRegressor']

scores = pd.DataFrame([model_names,score_train,score_test])
scores
```

```python
scores = scores.transpose() #creating new table
```

```python
scores.columns = [ 'Model','Training Set Accuracy','Testing set Accuracy']
```

```python
scores
```

```python
sns.barplot(data=scores, x='Model', y='Training Set Accuracy')
plt.title('Accuracy for each models', size=18)
plt.xticks(rotation=45,size=10)
plt.xlabel('Model',size=14)
plt.ylabel('Trainig Set Accuracy',size=14)
```

```python
sns.barplot(data=scores, x='Model', y='Testing set Accuracy')
plt.title('Accuracy for each models', size=18)
plt.xticks(rotation=45,size=10)
plt.xlabel('Model',size=14)
plt.ylabel('Testing Set Accuracy',size=14)
```

```python
#evaluating model using cross validation(to test the ability of the model)
from sklearn.model_selection import cross_val_score
xgb_score=cross_val_score(xgb,xtest,ytest,cv=4) 
print("Model accuracy is = {}".format(xgb_score.mean()*100))
```
